## Data Visualizations

```python
import pandas as pd

# Load the dataset
file = "Sleep_health_and_lifestyle_dataset.csv"
df = pd.read_csv(file) #read CSV into Python as a DataFrame

df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None') # Fill NaN with 'None'
df['Has Sleep Disorder'] = df['Sleep Disorder'].apply(lambda x: 0 if x == 'None' else 1) # Create boolean "Has Sleep Disorder" col 
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int) # split BP into systolic and diastolic
df = df.drop(['Person ID', 'Blood Pressure'], axis=1)
df # draw table
```

```python
# Pie Chart
df.groupby('Sleep Disorder').size().plot(kind='pie',colors = ['purple', 'lightblue', 'red'])
```

```python
# Box Plots
df.plot(kind='box', by = 'Sleep Disorder', column = 'Age')
df.plot(kind='box', by = 'Sleep Disorder', column = 'Heart Rate')
df.plot(kind='box', by = 'Sleep Disorder', column = 'Sleep Duration')
df.plot(kind='box', by = 'Sleep Disorder', column = 'Daily Steps')
```

```python
# Scatter Plots
df.plot(kind='scatter', x = 'Heart Rate', y = 'Quality of Sleep')
```
